```python
import yfinance as yf
import time

tickers = ['AAPL', 'MSFT', 'GOOG']
data = {}

for t in tickers:
    success = False
    while not success:
        try:
            df = yf.download(t, start="2000-01-01", end="2024-12-31")["Adj Close"]
            data[t] = df
            success = True
        except Exception as e:
            print(f"Retrying {t} after 60 seconds due to {e}")
            time.sleep(60)

import pandas as pd
prices = pd.concat(data, axis=1)
prices.to_csv("adj_close.csv")
```

```python
import pandas as pd
import numpy as np

# Read predicted returns
sel = pd.read_csv("stock_selected.csv")

# Optional: filter by date
date = "2011-06-01"
sel_date = sel[sel["trade_date"] == date].copy()

# Load price data for the same stocks
# Suppose you have 'adj_close.csv' with tickers (or gvkeys) as columns
# Rows = dates, Columns = gvkeys
prices = pd.read_csv("adj_close.csv", index_col=0, parse_dates=True)

# Match gvkeys that exist in both
valid_ids = [str(g) for g in sel_date["gvkey"].astype(str) if str(g) in prices.columns]
prices = prices[valid_ids]

# Compute returns and covariance
returns = prices.pct_change().dropna()
Sigma = returns.cov().values

# Expected returns (from your file)
mu = sel_date.set_index("gvkey").loc[valid_ids, "predicted_return"].values

# Compute Minimum-Variance weights
inv_Sigma = np.linalg.inv(Sigma)
ones = np.ones(len(mu))
w_minvar = inv_Sigma @ ones / (ones.T @ inv_Sigma @ ones)

# Compute portfolio statistics
mu_p = w_minvar @ mu
sigma_p = np.sqrt(w_minvar @ Sigma @ w_minvar)

# Display results
print(f"Trade date: {date}")
for name, w in zip(valid_ids, w_minvar):
    print(f"gvkey {name}: {w:.4f}")
print(f"\nExpected portfolio return: {mu_p:.4%}")
print(f"Portfolio volatility: {sigma_p:.4%}")
```
